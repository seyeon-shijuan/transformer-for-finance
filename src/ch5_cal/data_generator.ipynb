```python
import pandas as pd

# 데이터 프레임 생성
data1 = {
    '삼성전자': [1, 2, 3],
    'LG에너지솔루션': [4, 5, 6],
    'SK하이닉스': [7, 8, 9],
    '삼성바이오로직스': [10, 11, 12],
    '삼성전자우': [13, 14, 15]
}

data2 = {
    '삼성전자_p': [16, 17, 18],
    '에너지솔루션_p': [19, 20, 21],
    'SK하이닉스_p': [22, 23, 24],
    '삼성바이오로직스_p': [25, 26, 27],
    '삼성전자우_p': [28, 29, 30]
}

index = ['2023-11-01', '2023-11-02', '2023-11-03']

df1 = pd.DataFrame(data1, index=index)
df2 = pd.DataFrame(data2, index=index)

# 컬럼 순서 조정 및 컬럼 이름 변경
df = pd.merge(df1, df2, left_index=True, right_index=True)
df.columns = [
    '삼성전자', '삼성전자_p',
    'LG에너지솔루션', '에너지솔루션_p',
    'SK하이닉스', 'SK하이닉스_p',
    '삼성바이오로직스', '삼성바이오로직스_p',
    '삼성전자우', '삼성전자우_p'
]

# 결과 확인
df
```
